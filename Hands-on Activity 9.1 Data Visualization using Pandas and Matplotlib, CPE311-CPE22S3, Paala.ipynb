5. Using matplotlib and pandas, create two subplots side-by-side showing the effect that after-hours trading has had on Facebook's stock price:
   
   - The first subplot will contain a line plot of the daily difference between that day's opening price and the prior day's closing price (be sure to review the Time series section of Aggregating Pandas DataFrames for an easy way to do this).

    - The second subplot will be a bar plot showing the net effect this had monthly, using resample().

    - Bonus #1: Color the bars according to whether they are gains in the stock price (green) or drops in the stock price (red).

    - Bonus #2: Modify the x-axis of the bar plot to show the threeletter abbreviation for the month.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 10), constrained_layout=True)
change = fb.assign(
    difference = lambda x: x.open - x.close.shift() 
    # The shift function makes it so that there is a lagged data.
    # The price change in which it gets closed one day right up to the open following day is called after hours trading.
)

monthly = change.difference.resample('ME').sum()

bar_colors = ['green' if val >= 0 else 'red' for val in monthly] 
# A string will return, 'green' string if the value is greater or equal than zero.
# else, it will return 'red' string for each value in monthly.
# The strings work in the color argument.
axes[0].plot(change.index, change.difference, color='purple')
axes[0].set_title('Daily Difference Between Open and Prior Close Price')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Price Change')

axes[1].bar(monthly.index.strftime('%b'), monthly.values, color=bar_colors)
# strftime is a function where it takes the datetime type into a string.
# passing the %b, will result of it abbrevating the month name.
axes[1].set_title("Monthly Net Effect")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Net Change")
axes[1].grid(True, axis='y')
```
